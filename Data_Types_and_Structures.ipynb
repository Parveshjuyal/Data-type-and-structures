# Q4. Describe how dictionaries store data

    - Dictionaries (dict) in Python store data as key-value pairs, providing
      fast lookups, insertions, and deletions. Here's how they work:

      Each entry consists of a unique key mapped to a value.

# Keys must be immutable (e.g., str, int, tuple).

      Values can be any data type (e.g., int, list, another dict).

Example:

in python
      student = {
     "name": "Alice",  # key="name", value="Alice"  
     "age": 25,        # key="age", value=25  
     "courses": ["Math", "Physics"]  # key="courses", value=list  
}
  2. Underlying Implementation (Hash Tables)
     Dictionaries use hash tables for efficiency:

     Hashing the Key

     Python computes a hash (a unique integer) for the key using hash().

# Example: hash("name") → some fixed integer.

     Storing in Memory (Buckets)

     The hash determines the memory location (bucket) where the key-value pair is stored.

     Collisions (two keys with the same hash) are handled internally.

     Fast Lookup (O(1) Average Time)

     When accessing student["name"], Python:

     Recomputes the hash of "name".

     Directly jumps to the correct memory bucket.
     Returns the value.
     
     3. Mutable & Dynamic
     Dictionaries are mutable (can be modified after creation).
     You can add, remove, or update key-value pairs.
Example:
 in python
     student["grade"] = "A"      # Adds new key-value  
     student["age"] = 26         # Updates existing key  
     del student["courses"]      # Removes a key  
   4. Order Preservation (Python 3.7+)
     Since Python 3.7, dictionaries maintain insertion order (like collections.OrderedDict).
     Iterating over a dict returns keys in the order they were added.
Example:

in python
     d = {"a": 1, "b": 2, "c": 3}  
     print(list(d.keys()))  # Output: ["a", "b", "c"] (guaranteed order)

# Q8. What is a hash table, and how does it relate to dictionaries in Python3

    - A hash table is like a magic bookshelf where you can find things super
      fast. Here’s how it works:

# Keys → Magic Labels

      Every item (value) has a unique label (key), like "name" or "age".

      Python uses a hash function to turn the key into a number (like hash("name") = 12345).

 Numbers → Shelf Addresses

      That number tells Python exactly where to store the value in memory.

      No searching—just go straight to the right spot!

Fast Lookups

      Need person["name"]? Python:

      Computes hash("name") → 12345.

      Checks shelf 12345 → instantly finds "Alice".

      How Dictionaries Use Hash Tables
      In Python, dictionaries (dict) are built on hash tables!

      Feature	How It Works in Dictionaries

      Keys	Must be immutable (e.g., str, int, tuple).
      Values	Can be anything (lists, numbers, etc.).
      Speed	O(1) time for lookups (same speed no matter how big the dict is!).
      Collisions	Rare, but Python handles them automatically.

Example:

in python
      person = {"name": "Alice", "age": 25}  # Uses a hash table internally!  
      print(person["name"])  # Instantly finds "Alice" using `hash("name")`.

# Q19. Why are dictionaries considered unordered, and how does that affect data
       retrieval ?
      
    - Why Dictionaries Are "Unordered"
      1. The Myth of Unordered Dictionaries
      Before Python 3.7: Dictionaries were unordered (keys could appear in any sequence).

      Since Python 3.7: Dictionaries preserve insertion order (but this was an implementation detail until Python 3.8 made it official).

     2.Why They Used to Be Unordered
      Dictionaries use a hash table to store keys:

      Hashing: Keys are converted to random-looking numbers (hashes).

      Memory Placement: The hash determines where the key-value pair is stored in memory.
      → This led to seemingly "random" order when looping.

     3. How Order Preservation Now Affects You
      Insertion Order = Loop Order (since Python 3.7+):

in python
      d = {"a": 1, "b": 2, "c": 3}  
      print(list(d.keys()))  # Always ["a", "b", "c"]  
      But... Don’t rely on order for logic! Dictionaries are still optimized for key-based lookups, not ordering.

# Q20.Explain the difference between a list and a dictionary in terms of data
      retrieval.
    
    - How lists and dictionaries differ for data retrieval:
    1. Lists: Sequential Access (Like a Tape)
      Retrieval by Index:

      Access items by position (e.g., my_list[0]).

      Time Complexity: O(1) (instant for known positions).

Searching for a Value:

      Use in or loops (e.g., if "apple" in my_list).

      Time Complexity: O(n) (checks every item until found → slow for large lists).

Example:

in python
     fruits = ["apple", "banana", "orange"]  
     print(fruits[1])          # "banana" (fast)  
     print("pear" in fruits)   # False (must scan entire list)  
   2. Dictionaries: Direct Access (Like a Labeled Filing Cabinet)

Retrieval by Key:

     Access values using a key (e.g., my_dict["name"]).

     Complexity: O(1) (instant, uses hashing).

Checking for a Key:

     checks keys only (e.g., if "age" in my_dict).

     Time Complexity: O(1) (blazing fast).

Example:

python
     person = {"name": "Alice", "age": 30}  
     print(person["age"])      # 30 (instant)
